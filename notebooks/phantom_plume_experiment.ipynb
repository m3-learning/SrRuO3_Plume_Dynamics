```python
import numpy as np
import h5py
from matplotlib import pyplot as plt
```

```python
imgs = []
indexes = []
with h5py.File('../../SrRuO3_Plume_Dynamics/data/Plumes/plume_recordings/YG066_YichenGuo_09112024.h5', 'r') as f:
    print(f.keys())
    print(f['PLD_Plumes']['1-SrRuO3'].shape)
    for i in range(len(f['PLD_Plumes']['1-SrRuO3'])):
        img = f['PLD_Plumes']['1-SrRuO3'][i]
print(len(indexes))
print(len(imgs))
```

```python
imgs_sample = imgs[0:6]
indexes_sample = indexes[0:6]
fig, axes = plt.subplots(1, 6, figsize=(8, 2))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(imgs_sample[i])
    ax.set_title(f't = {indexes_sample[i]/2000:.4f} s')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
imgs = []
indexes = []
with h5py.File('../../SrRuO3_Plume_Dynamics/data/Plumes/plume_recordings/test_phantom_plume.h5', 'r') as f:
    print(f.keys())
    print(f['test1'].shape)
    for i in range(len(f['test1'])):
        img = f['test1'][i]
        if np.mean(img) > 20:
            imgs.append(img)
            indexes.append(i)
print(len(indexes))
print(len(imgs))
```

```python
imgs_sample = imgs[0:6]
indexes_sample = indexes[0:6]
fig, axes = plt.subplots(1, 6, figsize=(8, 2))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(imgs_sample[i])
    ax.set_title(f't = {indexes_sample[i]/2000:.4f} s')
    ax.axis('off')
plt.tight_layout()
plt.show()

imgs_sample = imgs[6:11]
indexes_sample = indexes[6:11]

fig, axes = plt.subplots(1, 5, figsize=(6.7, 2))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(imgs_sample[i])
    ax.set_title(f't = {indexes_sample[i]/2000:.4f} s')
    ax.axis('off')
plt.tight_layout()
plt.show()

imgs_sample = imgs[11:17]
indexes_sample = indexes[11:17]
fig, axes = plt.subplots(1, 5, figsize=(6.7, 2))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(imgs_sample[i])
    ax.set_title(f't = {indexes_sample[i]/2000:.4f} s')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(8, 5, figsize=(8, 10))
for i, ax in enumerate(axes.flatten()):
    if i < len(imgs):
        ax.imshow(imgs[i])
        ax.set_title(f't = {indexes[i]/2000:.4f} s')
        ax.axis('off')
plt.tight_layout()
plt.show()
```
