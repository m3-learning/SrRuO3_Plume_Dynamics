```python
# ! pip install xlrd
# ! pip install openpyxl

import pandas as pd
import matplotlib.pyplot as plt

def plot_xrd(file, save_path, n=2, intensity_difference=500):
    data = pd.read_excel('./Hao_Guo_XRD_data/COF-5 03032021.xlsx', header=29)
    
    X = data['Angle']
    Y1 = data[' Intensity']
    
    if n == 2:
        Y2 = Y1+intensity_difference
    
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(111)

    plt.plot(X, Y1)
    plt.plot(X, Y2)

    plt.title('custom title')

    ax.set_xlabel(r"2$\Theta}$", )
    ax.set_ylabel('Intensity [a.u.]')

    plt.savefig(save_path, dpi=1200)
    plt.show()

file_name = './Hao_Guo_XRD_data/COF-5 03032021.xlsx' # xrd excel file path
save_path = './Hao_Guo_XRD_data/fig2.svg' # save image file

plot_xrd(file_name, save_path=save_path, 
         n=2, intensity_difference=500)
```
